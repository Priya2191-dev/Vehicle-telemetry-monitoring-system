### 1. Speed Monitoring
This module monitors vehicle speed in real time. It helps ensure the vehicle operates within safe limits. Speed data is used for analytics and anomaly detection. It also feeds into braking and telemetry system.

```python
speed_data = [40, 50, 60, 120]

def monitor_speed(data):
    avg_speed = sum(data)/len(data)
    max_speed = max(data)
    return avg_speed, max_speed

avg, max_val = monitor_speed(speed_data)
print("Average Speed:", avg)
print("Max Speed:", max_val)
```

### 2. Engine Temperature Tracking

Tracks engine temperature to prevent overheating. Maintains safe thermal limits for engine health. Used in predictive maintainance systems. Triggers alerts when temperature exceeds threshold.

```python
temps = [70, 75, 80, 105]

def check_temp(data):
    return ["High" if t > 100 else "Normal" for t in data]

status = check_temp(temps)
print(status)
```

### 3. Brake Pressure Analysis

Analyzes braking force applied to the vehicle. Ensures optimal stopping performance. Helps simulate braking scenarios. Useful for safety validation systems.

```python
pressures = [10, 20, 30]

def brake_analysis(data):
    return [p * 0.2 for p in data]

result = brake_analysis(pressures)
print("Brake Impact:", result)
```

### 4. Telemetry Visualisation

Visualises telemetry data like speed and pressure. Helps to understand system behaviour. Used for debugging and reporting. Supports dashboard style monitoring.

```python
import matplotlib.pyplot as plt

speed = [10, 20, 30]
pressure = [5, 10, 15]

plt.plot(speed)
plt.plot(pressure)
plt.show()
```

### 5. AI Anomaly Detection

Detects abnormal patterns in telemetry data. Uses stastical deviation from mean values. Helps identify unsafe driving conditions. Foundation for AI-driven monitoring system.

```python
import numpy as np

data = [50, 55, 60, 120]

mean = np.mean(data)
anomaly = any(abs(x - mean) > 30 for x in data)

print("Anomaly Detected:", anomaly)
```

### 6. REST API using FastAPI

Exposes telemetry data via REST APIs. Allows integration with external systems. Supports real-time data access. Built using lightweight FastAPI framework

```python
from fastapi import FastAPI

app = FastAPI()

@app.get("/speed")
def get_speed():
    return {"speed": 60}

@app.get("/status")
def status():
    return {"status": "OK"}
```
